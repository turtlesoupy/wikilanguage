```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import pandas_helper
from pandas_helper import Concepts, Countries
pd.set_option('display.max_rows', 100)
n_display = 25
```

# Load in Database

```python
df = pandas_helper.load_data("data/wikilanguage.tsv")
```

# Most culturally relevant museums in the world

```python
df.wl.instance_of(Concepts.MUSEUM).wl.top_ranked("enwiki", n_display)
```

# ... but what about France

```python
df.wl.instance_of(Concepts.MUSEUM).wl.top_ranked("frwiki", n_display)
```

# How about specific to France?

```python
df.wl.instance_of(Concepts.MUSEUM).wl.kl_divergence(base_wiki="enwiki", target_wiki="frwiki").nlargest(n_display, "kl_divergence")[[
    "sample_label", "kl_divergence", "kl_relative_to_max"
]]
```

# Or those overrated by English speakers?

```python
df.wl.instance_of(Concepts.MUSEUM).wl.kl_divergence(base_wiki="frwiki", target_wiki="enwiki").nlargest(n_display, "kl_divergence")[[
    "sample_label", "kl_divergence", "kl_relative_to_max"
]]
```

# Near Palo Alto?

```python
palo_alto = df.wl.resolve("Palo Alto, California", col="enwiki_title")
museums = df.wl.instance_of(Concepts.MUSEUM)
palo_alto.wl.within_radius(museums, radius_km=100).wl.top_ranked("enwiki", n_display)
```

# Obscure ones near Palo Alto

```python
df.wl.instance_of(Concepts.MUSEUM).wl.kl_divergence(base_wiki="frwiki", target_wiki="enwiki").nlargest(n_display, "kl_divergence")[[
    "sample_label", "kl_divergence", "kl_relative_to_max"
]]palo_alto = df.wl.resolve("Palo Alto, California", col="enwiki_title")
museums = df.wl.instance_of(Concepts.MUSEUM)
palo_alto.wl.within_radius(museums, radius_km=100).wl.top_ranked("enwiki", n_display, desc=False)
```
